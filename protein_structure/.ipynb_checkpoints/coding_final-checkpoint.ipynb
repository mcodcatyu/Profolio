```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
```

```python
def combine (data):
    """
        Put separated [x],[y],[z] value of each amino acidsinto together 

        Args:
            data(pandas.core.frame.DataFrame): The seperated xyz position of 8 amino acids

        Returns:
            position(numpy.ndarray): [x,y,z] values of 8 amino acids
    """
    number = np.arange(0, 6000)
    name = ['Arg1', 'Ala1', 'His1','Pro1','Gly1','Gly2','Thr1','Ala2']
    position = np.zeros([6000,8,3]) # to store the x,y,z coordinates
    
    n = 0
    # gathering xyz coordinates into an array 
    for i in name:
        for j in range(0, 6000):
            x = data[f'{i}_x'][j]
            y = data[f'{i}_y'][j]
            z = data[f'{i}_z'][j]
            total = [x,y,z]
            position[j,n] = total
        n += 1
    return position
    
```

```python
def distance_calculate (position):
    """
        Calculate distances between the pairs of amino acids

        Args:
            position(numpy.ndarray) : [x,y,z] values of 8 amino acids

        Returns:
            dis_ary(numpy.ndarray) : stores the distance values between pairs of amino acids
    """
    first_ele = position[:, :4, :]
    second_ele = position[:, 4:, :]
    final_position = second_ele-first_ele
    dis_ary = np.zeros([6000,4])
    

    # Calculate the distances between amino acids, store the value in an array
    for i in range(0,4):
        for j in range(0,6000):
            minus_position = final_position[j, i]
            distance = np.linalg.norm(minus_position)
            dis_ary[j,i] = distance

    return dis_ary
```

```python
def plotting (df):
    """
        Prints the histogram to show the distribution of distances 

        Args:  df(pandas.core.frame.DataFrame) : stores the distance values between pairs of amino acids
    """
        
    fig, ax = plt.subplots(figsize=(10,6))
    
    ax.hist(df['A1G1'],  alpha = 0.5, label = 'Arg1–Gly1')
    ax.hist(df['A1G2'],  alpha = 0.5, label = 'Ala1–Gly2')
    ax.hist(df['H1T1'],  alpha = 0.5, label = 'His1–Thr1')
    ax.hist(df['P1A2'],  alpha = 0.5, label = 'Pro1–Ala2')
    
    ax.set_xlabel('Distance(Ångstrom)')
    ax.set_ylabel('Numbers')
    ax.set_xlim(4,12) 
    ax.set_ylim(0,2000)
    ax.legend()
    ax.grid(True)
    ax.set_title('Distances of 4 pairs')
    
    plt.show()
```

```python
nodrug_data = pd.read_csv('no_drug.csv')
drug_data = pd.read_csv('drug.csv')

#data combine
nodrug_pos = combine(nodrug_data)
drug_pos = combine(drug_data)

#calculate distance
nodrug_ary = distance_calculate(drug_pos) # ni tmd 這邊放相反了啦
drug_ary = distance_calculate(nodrug_pos)

nodrug_df = pd.DataFrame(nodrug_ary, columns=['A1G1', 'A1G2', 'H1T1','P1A2'], index=np.arange(1,6001))
drug_df = pd.DataFrame(drug_ary, columns=['A1G1', 'A1G2', 'H1T1','P1A2'], index=np.arange(1,6001))

#Plotting
plotting(nodrug_df)
plotting(drug_df)

```

```python
# no drug pca
pca_nodrug = PCA(n_components = 2)
pca_nodrug.fit(nodrug_df)
nodrug_pca = pca_nodrug.transform(nodrug_df)
no_pc1 = nodrug_pca[:, 0]
no_pc2 = nodrug_pca[:, 1]

print("No Drug \n", "The components:","\n", pca_nodrug.components_, '\n')
print("The explained_variance_: ", pca_nodrug.explained_variance_)
print("The explained_variance_ratio:", pca_nodrug.explained_variance_ratio_,'\n')

# drug pca
pca_drug = PCA(n_components = 2)
pca_drug.fit(drug_df)
drug_pca = pca_drug.transform(drug_df)

pc1 = drug_pca[:, 0]
pc2 = drug_pca[:, 1]

print("Drug \n", "The components:","\n", pca_drug.components_, '\n')
print("The explained_variance_: ", pca_drug.explained_variance_)
print("The explained_variance_ratio:", pca_drug.explained_variance_ratio_)
```

```python
# Plotting 
fig, ax = plt.subplots(figsize=(6,6))
ax.scatter(no_pc1, no_pc2, marker = '*', alpha=0.8, label='No drug')
ax.scatter(pc1, pc2, marker = 'o',  alpha=0.5, label = 'With Drug')
ax.axis('equal')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.legend()
plt.show()
```

```python
def mean_std_max_min(dis_df):
    """
        Calculate mean, std, max and min values

        Arg:
            dis_df(pandas.core.frame.DataFrame): stores the distance values between pairs of amino acids
        Returns:
             total(pandas.core.frame.DataFrame): stores the mean, std, maximum and minimum value of 4  amino acid pairs
    """
    msmm = np.zeros([4,4])
    
    #calculate mean
    for i,j in zip(dis_df.columns,range(0,4)):
        msmm[0,j] = dis_df[i].mean()
    #calculate std
    for i,j in zip(dis_df.columns,range(0,4)):
        msmm[1,j] = dis_df[i].std()   
    #calculate max
    for i,j in zip(dis_df.columns,range(0,4)):
        msmm[2,j] = dis_df[i].max()
    #calculate min
    for i,j in zip(dis_df.columns,range(0,4)):
        msmm[3,j] = dis_df[i].min()

    total =  pd.DataFrame(msmm, columns=(['Arg1–Gly1', 'Ala1–Gly2', ' His1–Thr1','Pro1–Ala2']), index=['mean', 'std','max', 'min'])
    return total
```

```python
mean_std_max_min(nodrug_df)
```

```python
mean_std_max_min(drug_df)
```

```python
def kmeans_plot(drug_df, n_components):
    """
        Apply k-means algorithm to cluster data and print plot

        Arg:
            drug_df(pandas.core.frame.DataFrame): stores the distance values between pairs of amino acids with drug
            n_components(int): the component number for PCA and cluster number for k-means
    """
    pca_drug = PCA(n_components)
    pca_drug.fit(drug_df)
    drug_pca = pca_drug.transform(drug_df)
    
    X = drug_pca
    n_clusters = n_components
    kmeans = KMeans(n_clusters)
    kmeans.fit(X)
    labels = kmeans.labels_
    
    fig, ax = plt.subplots(figsize=(8,6))
    ax.scatter(X[:, 0], X[:, 1], c=labels)
    ax.scatter(X[0,0], X[0, 1], marker = 'o', label='first point')
    ax.scatter(X[-1,0], X[-1,1], marker = 's', label = 'last time point')
    ax.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color = 'red', marker='*', label = 'centre')
    ax.set_xlabel('PC1')
    ax.set_ylabel('PC2')
    ax.legend()
    
    plt.show()

```

```python
kmeans_plot(drug_df, n_components=2)
```

```python
kmeans_plot(drug_df, n_components=3)
```

```python
def gaussian_plot(drug_df, n_components):
    """
        Apply gaussian algorithm to cluster data and print plot

        Arg:
            drug_df(pandas.core.frame.DataFrame): stores the distance values between pairs of amino acids with drug
            n_components(int): the component number for PCA and cluster number for gaussian
    """
    pca_drug = PCA(n_components)
    pca_drug.fit(drug_df)
    drug_pca = pca_drug.transform(drug_df)
    
    X = drug_pca
    gmm = GaussianMixture(n_components).fit(X)
    labels = gmm.predict(X)
    probs = gmm.predict_proba(X)
    size = 10 * probs.max(axis=1)**2
    
    
    fig, ax = plt.subplots(figsize=(8,6))
    ax.scatter(X[:, 0], X[:, 1], c=labels, s = size)
    ax.scatter(X[0,0], X[0, 1], marker = 'o')
    ax.scatter(X[-1,0], X[-1,1], marker = 's')
    ax.axis('equal')
    ax.set_xlabel('PC1')
    ax.set_ylabel('PC2')
    plt.show()
```

```python
gaussian_plot(drug_df, n_components=2)
```

```python
gaussian_plot(drug_df, n_components=3)
```
